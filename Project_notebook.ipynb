```python
def cleaning_outlier(x):
    date = x.Date
    mean = rolling_mean.loc[rolling_mean.index == date]
    std = rolling_std.loc[rolling_std.index == date]
    if mean.isnull().sum().sum() != len(mean.columns):
        if x.Price > mean.Price.iloc[0] + 2 * std.Price.iloc[0]:
            x.Price = np.nan
        elif x.Price < mean.Price.iloc[0] - 2 * std.Price.iloc[0]:
            x.Price = np.nan
    return x
```
